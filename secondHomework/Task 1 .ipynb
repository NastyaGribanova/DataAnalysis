3. Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?

```python
abs10 = data[data['absences'] > 10]
other = data[data['absences'] <= 10]
mark_abs10 = abs10['G3'].mean()
mark_other = other['G3'].mean()
if mark_abs10 < mark_other:
    print('Да, это правда')
else: print('Нет')
```
